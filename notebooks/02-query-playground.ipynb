# 02 – Query playground

Use this notebook to run questions against the ingested index and inspect retrieved chunks.

**Before running**, complete notebook `01-ingest-playground.ipynb` at least once so the vector store is populated.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))

from rag import config
from rag.query import ask
```

```python
QUESTION = "What does the NPPF say about green belt?"

result = ask(QUESTION, top_k=5)
print(result.answer)
```

```python
# Inspect citations
for i, c in enumerate(result.citations, start=1):
    print(f'[{i}] {c.category} / {c.source} (chunk {c.chunk_index})')
    print(f'    {c.excerpt}')
    print()
```

```python
# Filter by category – e.g. national policy only
result_national = ask(QUESTION, top_k=5, category_filter='national')
print(result_national.answer)
```

```python
# Batch a list of questions
questions = [
    "What is a material consideration in planning decisions?",
    "When can development be carried out without planning permission?",
    "What is the definition of 'development' under the TCPA 1990?",
]

for q in questions:
    r = ask(q, top_k=3)
    print(f'Q: {q}')
    print(f'A: {r.answer[:300]}...')
    print()
```
